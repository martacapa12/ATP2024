```python
def menu():
    print("""
    (1): Criar uma turma;
    (2): Inserir um aluno na turma;
    (3): Listar a turma;
    (4): Consultar um aluno por id;
    (5): Guardar a turma em ficheiro;
    (6): Carregar uma turma dum ficheiro;
    (0): Sair da aplicação
    """)

def criarTurma():
    turma1 = []


def inserirAluno(turma, nome, id, notaTPC, notaProj, notaTeste):
    notas = [notaTPC, notaProj, notaTeste]
    turma.append((nome, id, notas))


def listarTurma(turma):
    for aluno in turma:
        print(aluno)
    
    
def consultarId(turma, id):
    for aluno in turma:
        if aluno[1] == id:
            print(aluno)
        else:
            print(f"O aluno {id} não existe.")
    
    
def guardarTurma(turma, fnome):
    file = open(fnome,"w")            
    for aluno in turma:
        nome, id, notas = aluno
        notaTPC, notaProj, notaTeste = notas
        file.write(str(nome) + "|" + str(id) + "|" + str(notaTPC) + "|" + str(notaProj) + "|" + str(notaTeste))
        file.write("\n")
    file.close()          
    
def carregarTurma(fnome):          
    turma1 = []
    file = open(fnome,"r")
    for line in file:
        elementos = line.split("|")
        nome, id, notaTPC, notaProj, notaTeste = elementos 
        notas = [int(notaTPC), int(notaProj), int(notaTeste)] 
        turma1.append((nome, id, notas))
    file.close()
    return turma1 


turma1 = []

cond = True
while cond:
    menu()
    opcao = int(input("Indique qual operação quer realizar: "))
    if opcao == 1:
        criarTurma()
    elif opcao == 2:
        nomeAluno = input("Introduza o nome do aluno: ")
        idAluno = input("Introduza o id do aluno: ")
        notaTPCAluno = int(input("Introduza a nota que o aluno obteve no TPC: "))
        notaProjAluno = int(input("Introduzir a nota do aluno no projeto: "))
        notaTesteAluno = int(input("Introduzir a nota do aluno no teste: "))
        inserirAluno(turma1, nomeAluno, idAluno, notaTPCAluno, notaProjAluno, notaTesteAluno)
    elif opcao == 3:
        listarTurma(turma1)
    elif opcao == 4:
        idAluno = input("Introduza o id do aluno: ")
        consultarId(idAluno)
    elif opcao == 5:
        guardarTurma(turma1, "turma.txt")
    elif opcao == 6:
        print(carregarTurma("turma.txt"))
    elif opcao == 0:
        cond = False
    elif opcao > 6 or opcao < 0:
        print("Essa opção não existe.")

menu()
```
